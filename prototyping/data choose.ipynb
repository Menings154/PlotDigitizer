```python
import matplotlib.pyplot as plt
import matplotlib.patches as pt
import cv2
```

```python
img_path = r"C:\Users\Benja\Code\Python\PlotDigitizer\data\plot digitizer test.PNG"

img = cv2.imread(img_path)
```
